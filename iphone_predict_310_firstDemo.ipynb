```python
import pandas as pd
import numpy as np
data=pd.read_excel('iPhone6s.xlsx')
```

```python
#展示数据
data.columns=['price','where','fix','storage','color','repaire','phone_look','screen_look','screen_show','id']
data.head()
```

```python
#数据格式转换
data1=pd.get_dummies(data[['price','where','fix','repaire','phone_look','screen_look','screen_show','id']])
data1['storage']=data['storage']
data1.head()
```

```python
#去除128GB后面的GB
data1['storage']=data1['storage'].str.split('G').str[0]
data1.head()
```

```python
#数据分析
data1.isnull().sum()
```

```python
#数据统计性信息
data1.describe()
```

```python
#导入所需包
import time 
from sklearn.ensemble import RandomForestClassifier,RandomForestRegressor
from sklearn import metrics,cross_validation
from sklearn.metrics import mean_squared_error
#import scikitplot.plotters as skplt
from sklearn.model_selection import StratifiedKFold
import matplotlib.pylab as plt
from sklearn.grid_search import GridSearchCV
%matplotlib inline
```

```python
#提取标签数据
y=data1['price']
y.head()
```

```python
#删除无用特征
data1.drop(['price','id'],axis=1,inplace=True)
data1.columns.values
```

```python
data1.head()
```

```python
#转换数据类别
data1=np.array(data1)
```

```python
#分成训练和测试集两部分
from sklearn.model_selection import ShuffleSplit
ss=ShuffleSplit(n_splits=1,test_size=.1,random_state=0,train_size=.9)
ss.get_n_splits(data1)
print(ss)
for train_index,test_index in ss.split(data1,y):
    #print("Train Index:",train_index,",Test Index:",test_index)
    X_train,X_test=data1[train_index],data1[test_index]
    y_train,y_test=y[train_index],y[test_index]
```

```python
#构建随机森林回归模型
rf1=RandomForestRegressor(n_estimators=60,max_depth=15,max_features=19,oob_score=True,random_state=10)
%time rf1.fit(X_train,y_train)
y_predprob1=rf1.predict(X_test)#预测数据
print(rf1.oob_score_,mean_squared_error(y_test, y_predprob1))#输出袋外准确率，泛化能力体现，输出均方误差
#袋外准确率在0.994，均方误差为600.14
```

```python
#模型比较
y_test.std()#待预测数据目标值 标准差
```

```python
np.sqrt(mean_squared_error(y_test, y_predprob1))#预测结果值的标准差
```
